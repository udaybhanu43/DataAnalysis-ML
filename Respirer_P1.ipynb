```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data
data = pd.read_excel('/content/drive/MyDrive/Respirer Ass/cons_data_all_res.xlsx')
```

```python
data
```

```python
data['Time'] = pd.to_datetime(data['Time'])
data.set_index('Time', inplace=True)
```

```python
data
```

```python
data['Month'] = data.index.month
data['Day'] = data.index.day
data['Hour'] = data.index.hour
data['Weekday'] = data.index.weekday
data
```

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
data['Season'] = label_encoder.fit_transform(data['season'])

data
```

##Seasonal Plots

```python
plt.figure(figsize=(12, 8))
sns.lineplot(x=data.index, y=data['r1'], hue=data['Season'], palette='tab10')
plt.title('Meter Reading r1 by Season')
plt.xlabel('Time')
plt.ylabel('Meter Reading')
plt.legend(title='Season')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

meters = ['r1', 'r2', 'r3', 'r4', 'r5', 'r6', 'r7', 'r8', 'r9', 'r10', 'r11', 'r12']
n_meters = len(meters)
n_cols = 4
n_rows = 3

colors = sns.color_palette("Set2", len(data['season'].unique()))

fig, axes = plt.subplots(n_rows, n_cols, figsize=(20, 15))

axes = axes.flatten()

for i, meter in enumerate(meters):
    avg_data = data.groupby('season')[meter].mean()
    avg_data.plot(kind='bar', ax=axes[i], color=colors, edgecolor='black')

    axes[i].set_title(f'Seasonal Meter Reading: {meter}', fontsize=14, fontweight='bold')
    axes[i].set_xlabel('Season', fontsize=12)
    axes[i].set_ylabel('Average Reading', fontsize=12)
    axes[i].set_xticks(range(len(data['season'].unique())))
    axes[i].set_xticklabels(data['season'].unique(), rotation=45, ha='right')
    axes[i].tick_params(axis='x', labelsize=10)
    axes[i].tick_params(axis='y', labelsize=10)
    axes[i].grid(True, linestyle='--', alpha=0.7)

plt.tight_layout()
plt.subplots_adjust(top=0.95)
fig.suptitle('Seasonal Meter Readings', fontsize=16, fontweight='bold')
plt.show()
```

```python
colors = sns.color_palette("Set2", len(data['season'].unique()))

fig, axes = plt.subplots(n_rows, n_cols, figsize=(20, 15))

axes = axes.flatten()

for i, meter in enumerate(meters):
    sns.boxplot(x='season', y=meter, data=data, ax=axes[i], palette=colors)

    axes[i].set_title(f'Seasonal Meter Reading: {meter}', fontsize=14, fontweight='bold')
    axes[i].set_xlabel('Season', fontsize=12)
    axes[i].set_ylabel('Reading', fontsize=12)
    axes[i].tick_params(axis='x', labelsize=10, rotation=45)
    axes[i].tick_params(axis='y', labelsize=10)
    axes[i].grid(True, linestyle='--', alpha=0.7)


plt.tight_layout()
plt.subplots_adjust(top=0.95)
fig.suptitle('Seasonal Meter Readings Boxplot', fontsize=16, fontweight='bold')
plt.show()
```

##Monthly Plots

```python
colors = sns.color_palette("husl", len(meters))
fig, ax = plt.subplots(figsize=(15, 10))

for i, meter in enumerate(meters):
    avg_data = data.groupby('month_name')[meter].mean()
    avg_data.plot(kind='line', marker='o', ax=ax, color=colors[i], linestyle='-', linewidth=2, markersize=6, label=meter)

ax.set_title('Monthly Meter Readings', fontsize=16, fontweight='bold')
ax.set_xlabel('Month', fontsize=12)
ax.set_ylabel('Average Reading', fontsize=12)
ax.set_xticks(range(0, 12))
ax.set_xticklabels(range(0, 12), rotation=45, ha='right')
ax.tick_params(axis='x', labelsize=10)
ax.tick_params(axis='y', labelsize=10)
ax.grid(True, linestyle='--', alpha=0.7)
ax.legend(title='Meter', fontsize=10, title_fontsize=12)

plt.tight_layout()
plt.subplots_adjust(top=0.9)
plt.show()
```

##Weekly Plots

```python
colors = sns.color_palette("Set2", 7)
fig, axes = plt.subplots(n_rows, n_cols, figsize=(20, 15))
axes = axes.flatten()

for i, meter in enumerate(meters):
    avg_data = data.groupby('dow_date')[meter].mean()
    avg_data.plot(kind='bar', ax=axes[i], color=colors, edgecolor='black')

    axes[i].set_title(f'Daily Meter Reading: {meter}', fontsize=14, fontweight='bold')
    axes[i].set_xlabel('Day of Week', fontsize=12)
    axes[i].set_ylabel('Average Reading', fontsize=12)
    axes[i].set_xticks(range(7))
    axes[i].set_xticklabels(['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun'], rotation=0)
    axes[i].tick_params(axis='x', labelsize=10)
    axes[i].tick_params(axis='y', labelsize=10)
    axes[i].grid(True, linestyle='--', alpha=0.7)

plt.tight_layout()
plt.subplots_adjust(top=0.95)
fig.suptitle('Daily Meter Readings', fontsize=16, fontweight='bold')
sns.set_palette("Set1")
plt.show()
```

##hourly Plots

```python
colors = sns.color_palette("husl", 24)
fig, axes = plt.subplots(n_rows, n_cols, figsize=(20, 15))
axes = axes.flatten()

for i, meter in enumerate(meters):
    avg_data = data.groupby('hour')[meter].mean()
    avg_data.plot(kind='barh', ax=axes[i], color=colors)

    axes[i].set_title(f'Hourly Meter Reading: {meter}', fontsize=14, fontweight='bold')
    axes[i].set_ylabel('Hour of Day', fontsize=12)
    axes[i].set_xlabel('Average Reading', fontsize=12)
    axes[i].set_yticks(range(24))
    axes[i].set_yticklabels(range(24))
    axes[i].tick_params(axis='y', labelsize=10)
    axes[i].tick_params(axis='x', labelsize=10)
    axes[i].grid(True, linestyle='--', alpha=0.7)

plt.tight_layout()
plt.subplots_adjust(top=0.95)
fig.suptitle('Hourly Meter Readings', fontsize=16, fontweight='bold')
plt.show()
```
